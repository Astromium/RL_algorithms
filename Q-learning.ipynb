```python
import numpy as np
import gymnasium as gym
#import gym 
import matplotlib.pyplot as plt
from tqdm import tqdm
import random
```

```python
env = gym.make("FrozenLake-v1", map_name="8x8", is_slippery=False)
print(env.observation_space)
print(env.action_space)
```

```python
n_episodes = 10000
max_steps = 300
eps_start = 1.0
eps = 1.0
eps_end = 0.01
eps_decay_rate = 0.001
learning_rate = 0.1
gamma = 0.99
episodes_rewards = []
```

```python
def init_q(env):
    actions = env.action_space.n
    observations = env.observation_space.n
    return np.zeros((observations, actions))
```

```python
Q_table = init_q(env)
print(Q_table.shape)
```

```python
def train_q(env, eps, eps_start, eps_end, eps_decay_rate, gamma, learning_rate, n_episodes, max_steps):
    q_table = init_q(env)
    for episode in tqdm(range(n_episodes)):
        observation = env.reset()[0]
        rewards = 0.0
        for step in range(max_steps):
            if random.random() <= eps:
                action = env.action_space.sample()
            else:
                action = np.argmax(q_table[observation, :])
            new_observation, reward, terminated, truncated, _ = env.step(action)
            #print("new_observation : {new_observation}")
            rewards += reward
            # Update the Q-table
            q_table[observation, action] = q_table[observation, action] + learning_rate * (reward + gamma*np.max(q_table[new_observation, :]) - q_table[observation, action])
            observation = new_observation
            if terminated or truncated:
                episodes_rewards.append(rewards)
                break
        episodes_rewards.append(rewards)
        if episode % 1000 == 0:
            print(f"Episode {episode}, Reward : {rewards}")
        eps = eps_end + (eps_start - eps_end) * np.exp(-eps_decay_rate * episode)
    return q_table
```

```python
Q_table = train_q(env, eps, eps_start, eps_end, eps_decay_rate, gamma, learning_rate, n_episodes, max_steps)
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False, render_mode='human')
observation = env.reset()[0]
for i in range(10):
    env.render()
    action = np.argmax(Q_table[observation, :])
    new_obs, reward, terminated, truncated, _ = env.step(action)
    if terminated or truncated:
        observation = env.reset()[0]
        continue
    observation = new_obs

env.close()
```
